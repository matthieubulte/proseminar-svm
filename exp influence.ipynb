```julia
using PyPlot
pygui(true)
ion()
```

```julia
n = 100
x = linspace(-1, 1, n)
y = linspace(-1,1,n)

xgrid = repmat(x',n,1)
ygrid = repmat(y,1,n)

z = zeros(n,n)


p = .5*[ 1 1; -1 .2; .5 -1 ]
l = [1; 1; -1]
for i in 1:n
    for j in 1:n
        z[i:i,j:j] = 0
        for k=1:size(p,1)
            z[i:i,j:j] += l[k] * exp(-norm([x[i]; y[j]] - p[k,:])*5)
        end
    end
end

clf()
scatter3D(p[[1; 2], 2], p[[1; 2], 1], zeros(2), alpha=1, c="black", marker="x", s=50)
scatter3D(p[[3], 2], p[[3], 1], zeros(1),  edgecolors="black", marker="o", s=50, facecolors="none")
plot_surface(xgrid, ygrid, z, cmap=PyPlot.cm_get_cmap("gnuplot"), linewidth=0, antialiased=false, alpha=.6)

ax = plt[:gca](projection="3d")
ax[:_axis3don] = false


show()
```
